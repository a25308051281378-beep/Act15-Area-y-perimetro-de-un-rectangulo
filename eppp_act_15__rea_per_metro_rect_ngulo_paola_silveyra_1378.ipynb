#### Paola Silveyra NL = 50 Fecha 13/04/26
Desarrolla un programa en Python que funcione como una calculadora geométrica interactiva. El sistema debe mostrar un menú que permita al usuario calcular el área o el perímetro de un rectángulo. Cada opción solicitará los valores necesarios (base y altura). El programa se repetirá hasta que el usuario decida salir.

#### 📐 Fórmulas a utilizar
#### Área del rectángulo:
* A = base × altura
* Perímetro del rectángulo:
* P = 2 × (base + altura)

<img src="https://raw.githubusercontent.com/a25308051281378-beep/Act15-Area-y-perimetro-de-un-rectangulo/refs/heads/main/Formulas-del-perimetro-y-del-area-de-un-rectangulo.webp" width="500" height="600" alt="Texto descriptivo">

```python
# -----------------------------------------
# PROGRAMA: AREA Y PERIMETRO DEL RECTANGULO
# -----------------------------------------

# Funcion para calcular area
def area_rectangulo():
    print("\nCalculo del area del rectangulo")
    base = float(input("Ingresa la base: "))
    altura = float(input("Ingresa la altura: "))
    area = base * altura
    print("El area es:", round(area, 2))


# Funcion para calcular perimetro
def perimetro_rectangulo():
    print("\nCalculo del perimetro del rectangulo")
    base = float(input("Ingresa la base: "))
    altura = float(input("Ingresa la altura: "))
    perimetro = 2 * (base + altura)
    print("El perimetro es:", round(perimetro, 2))


# Funcion menu
def menu():
    while True:
        print("\n--- MENU ---")
        print("1. Calcular area del rectangulo")
        print("2. Calcular perimetro del rectangulo")
        print("3. Salir")

        opcion = input("Elige una opcion: ")

        if opcion == "1":
            area_rectangulo()

        elif opcion == "2":
            perimetro_rectangulo()

        elif opcion == "3":
            print("Programa terminado")
            break

        else:
            print("Opcion no valida")


# Programa principal
if __name__ == "__main__":
    menu()
```
